```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.manifold import LocallyLinearEmbedding
```

```python
wine_data = pd.read_csv("winequality-red.csv", delimiter = ";")
```

```python
wine_data
```

```python
embedding = LocallyLinearEmbedding(n_components=2)
wine_data_transformed = embedding.fit_transform(wine_data)
```

```python
plt.scatter(wine_data_transformed[:, 0], wine_data_transformed[:, 1], c = wine_data["quality"])
plt.colorbar(label = "Wine Quality")
plt.xlabel("LLE Component 1")
plt.ylabel("LLE Component 2")
plt.title("Local Linear Embedding with two Components")
```
